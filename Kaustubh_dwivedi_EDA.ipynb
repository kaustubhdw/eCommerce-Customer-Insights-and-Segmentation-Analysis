```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url1 = "https://drive.google.com/uc?export=download&id=1bu_--mo79VdUG9oin4ybfFGRUSXAe-WE"
df1 = pd.read_csv(url1)
url2 = "https://drive.google.com/uc?export=download&id=1IKuDizVapw-hyktwfpoAoaGtHtTNHfd0"
df2 = pd.read_csv(url2)
url3 = "https://drive.google.com/uc?export=download&id=1saEqdbBB-vuk2hxoAf4TzDEsykdKlzbF"
df3 = pd.read_csv(url3)
```

```python
print(df1.isnull().sum())
print(df2.isnull().sum())
print(df3.isnull().sum())
```

```python
region_counts = df1['Region'].value_counts()
region_counts.plot(kind='bar', title='Customer Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.show()
```

```python
average_price = df2.groupby('Category')['Price'].mean()
average_price.plot(kind='bar', title='Average Price of Products by Category')
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.show()
```

```python
df3['TransactionDate'] = pd.to_datetime(df3['TransactionDate'])
df3['YearMonth'] = df3['TransactionDate'].dt.to_period('M')
transactions_by_month = df3['YearMonth'].value_counts().sort_index()
transactions_by_month.plot(kind='line', title='Transaction Trends Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
top_products = df3['ProductID'].value_counts().head(10)
top_products.plot(kind='bar', title='Top-Selling Products')
plt.xlabel('Product ID')
plt.ylabel('Number of Transactions')
plt.show()
```
