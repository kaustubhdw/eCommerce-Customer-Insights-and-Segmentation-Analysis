```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
url1 = "https://drive.google.com/uc?export=download&id=1bu_--mo79VdUG9oin4ybfFGRUSXAe-WE"
df1 = pd.read_csv(url1)
url2 = "https://drive.google.com/uc?export=download&id=1IKuDizVapw-hyktwfpoAoaGtHtTNHfd0"
df2 = pd.read_csv(url2)
url3 = "https://drive.google.com/uc?export=download&id=1saEqdbBB-vuk2hxoAf4TzDEsykdKlzbF"
df3 = pd.read_csv(url3)
```

```python
df_merged = pd.merge(df3, df1, on='CustomerID')
df_merged = pd.merge(df_merged, df3, on='ProductID')
```

```python
df_merged
```

```python
customer_features = df_merged.groupby('CustomerID_x').agg({
    'TotalValue_x': 'sum',
    'TransactionID_x': 'count'
}).rename(columns={'TotalValue_x': 'TotalSpend', 'TransactionID_x': 'TransactionCount'})

df_customers = df1.rename(columns={'CustomerID': 'CustomerID_x'})

customer_data = pd.merge(df_customers, customer_features, on='CustomerID_x', how='left')

print(customer_data.head())
```

```python
customer_data.fillna(0, inplace=True)
print(customer_data.isnull().sum())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

```

```python
scaler = StandardScaler()
customer_data_scaled = scaler.fit_transform(customer_data[['TotalSpend', 'TransactionCount']])

similarity_matrix = cosine_similarity(customer_data_scaled)
```

```python
def get_top_n_similar_customers(customer_id, n=3):
    customer_idx = customer_data.index[customer_data['CustomerID_x'] == customer_id].tolist()[0]
    similarity_scores = similarity_matrix[customer_idx]
    top_n_indices = similarity_scores.argsort()[-n-1:-1][::-1]
    top_n_customers = customer_data.iloc[top_n_indices]['CustomerID_x']
    top_n_scores = similarity_scores[top_n_indices]
    return list(zip(top_n_customers, top_n_scores))

lookalikes = {customer_id: get_top_n_similar_customers(customer_id) for customer_id in customer_data['CustomerID_x'][:20]}
```

```python
lookalikes_df = pd.DataFrame([(k, [cust_id for cust_id, score in v], [score for cust_id, score in v]) for k, v in lookalikes.items()],
                             columns=['CustomerID_x', 'Lookalikes', 'Scores'])
lookalikes_df.to_csv('Lookalike.csv', index=False)

print(lookalikes_df.head())
```
